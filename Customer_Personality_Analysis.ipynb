## Handle Missing Values

```python
df.info()
```

#### The df.info() output will show you the columns, the number of non-null values, and their data types. This is where you'll first notice the missing values in the 'Income' column.

```python
mean_income = df['Income'].mean()
df['Income'].fillna(mean_income, inplace=True)
```

#### fillna() fills the missing values in the 'Income' column with the mean

## Remove Duplicate Rows

```python
print(f"Number of duplicate rows before removal: {df.duplicated().sum()}")
```

#### This is to verify if there are any duplicate rows in the dataset

```python
print(df.isnull().sum())
```

#### isnull() verifies that there are no more missing values in the 'Income' column
